```python
Ratings = pd.read_csv("ratings.csv")
```

```python
#Checking for NaN values
Ratings.apply(lambda x: x.isnull().sum(),axis=0)
```

```python
#Creating dataset with particular columns
Books_dataset = pd.DataFrame(Books, columns=['book_id', 'authors', 'title', 'average_rating'])
```

```python
#Dataset is sorted by book_id values
Books_dataset = Books_dataset.sort_values('book_id')
```

```python
Books_dataset['book_id']
```

```python
#Merging Books_dataset  and Rating and forming Books_newdata
Books_newdata = pd.merge(Books_dataset, Ratings, on='book_id')
```

```python
Books_newdata
```

```python
#Get rating of each book 
each_book_rating = pd.pivot_table(Books_newdata, index='user_id', values='rating', columns='title', fill_value=0)
```

```python
each_book_rating
```

```python
#Correlation of each book 
#Here T is used such that it will be applicable for books not id
book_corr = np.corrcoef(each_book_rating.T)
```

```python
book_corr
```

```python
#shape of array
book_corr.shape
```

```python
# To get sorted list of books based on correlation index values
book_list=  list(each_book_rating)
book_titles =[] 
for i in range(len(book_list)):
    book_titles.append(book_list[i])
```

```python
book_titles
```

```python
#Input book for similar recommendation
book = 'Harry Potter and the Prisoner of Azkaban (Harry Potter, #3)'
book_index = book_titles.index(book)
```

```python
corr_score = book_corr[book_index]
```

```python
#Based on the correlation score of index values similar books are recommended
condition = (corr_score >= 0.1)
```

```python
#These are the books "books recommended" based on the input book 'Harry Potter and the Prisoner of Azkaban (Harry Potter, #3)'
np.extract(condition, book_titles)
```
